**Let us try to code and extract from the html text the authors and make the file more readable to the user.**

```python
f = open("quotes.txt","w")

r = requests.get("https://quotes.toscrape.com/")
r = r.text
lines = r.split("\n")
for line in lines:
  if('<span class="text" itemprop="text">' in line):
    line = line.replace('<span class="text" itemprop="text">“',"").replace('”</span>',"").replace('&#39;',"'") # to make the lines more presentable
    line = line.strip()
    f.write(line+'\n')
  elif('<small class="author" itemprop="author">' in line):
    line = line.replace('<span>by <small class="author" itemprop="author">',"").replace('</small>',"")
    line = line.strip()
    f.write('by '+line+2*'\n')
f.close()
```

**However, what I am interested is to know all the quotes and their authors in the first five pages. One solution can be going through pages and copying their URL and appending it to the file. But that involves lot of manual work. Instead, we can modify the already written code and add a for loop to it.**

```python
f = open("quotes.txt","w")
for i in range(1,6):
  r = requests.get(f"https://quotes.toscrape.com/page/{i}/")
  r = r.text
  lines = r.split("\n")
  for line in lines:
    if('<span class="text" itemprop="text">' in line):
      line = line.replace('<span class="text" itemprop="text">“',"").replace('”</span>',"").replace('&#39;',"'")
      line = line.strip()
      f.write(line+'\n')
    elif('<small class="author" itemprop="author">' in line):
      line = line.replace('<span>by <small class="author" itemprop="author">',"").replace('</small>',"")
      line = line.strip()
      f.write('by '+line+2*'\n')
f.close()
```
